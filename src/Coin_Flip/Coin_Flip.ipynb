```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define the sequence of flips
flip_sequence = ["H", "H", "T", "H"]

# Initialize parameters for the prior Beta distribution
alpha = 1  # Start with Beta(1,1) prior, equivalent to uniform
beta_param = 1

# Set up the plot
plt.figure(figsize=(10, 8))

# Define theta range for plotting
theta = np.linspace(0, 1, 100)

# Loop through each flip and update the posterior
for i, flip in enumerate(flip_sequence):
    # Update alpha and beta based on the flip outcome
    if flip == "H":  # Head observed
        alpha += 1
    else:  # Tail observed
        beta_param += 1
    
    # Calculate the Beta distribution PDF with updated alpha and beta
    beta_pdf = beta.pdf(theta, alpha, beta_param)
    
    # Plot the updated Beta distribution
    plt.plot(theta, beta_pdf, label=f"After flip {i+1} ({flip}): Beta({alpha},{beta_param})")
    plt.fill_between(theta, beta_pdf, alpha=0.2)

#Labels
plt.title("Posterior Distributions of  After Each Coin Flip")
plt.xlabel(r"$\theta$")
plt.ylabel("Density")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
